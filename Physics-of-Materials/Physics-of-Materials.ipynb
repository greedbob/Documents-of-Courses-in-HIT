3 Wavefunction and its meaning.

A wave function is a mathematical tool used in quantum mechanics to describe the momentary states of subatomic particles, and it is expressed $Ψ(x,y,z,t)$ or $Ψ(\overrightarrow{r},t)$.

The value of the wave function is probability amplitude, and the square of its absolute value gives the probability distribution that the system will be in any possible state.

4 The solution of Schrodinger equation of the particles bound in one-dimensional potential wells.

The Schrodinger equation $\frac{-\hbar}{2m}\frac{\partial^{2}\Psi}{\partial x^{2}} + U\Psi = i\hbar\frac{\partial \Psi}{\partial t}$

For potential is not related with time and one-dimensional situation, $\Psi(x,t)=\psi(x)e^{\frac{-iEt}{\hbar}}$,

The Schrodinger equation is $-\frac{{\hbar}^2}{2m}\frac{\partial^{2}\psi}{\partial x^2} + U\psi = E\psi$

In the range of $x\leqslant 0$ or $x\geqslant a$, $U=\infty$, so $\psi=0$

In the range of $0\leqslant x\leqslant a$, $U=0$, $\frac{-\hbar}{2m}\frac{\partial^{2}\Psi}{\partial x^{2}}$

let $k^2=\frac{2mE}{\hbar^2}$, $\frac{\partial^2 \psi}{\partial x^2} + k^2\psi = 0$, thus $\psi(x) = A\cos{kx}+B\sin{kx}$

at point of $x=0$ and $x=a$, $\psi(x)=0$, thus \begin{equation}\left\{\begin{array}{lr}\psi(0)=A=0\\\psi(a)=B\sin{ka}=0 \end{array} \right.\end{equation}

thus $ka = n\pi$, $k=\frac{n\pi}{a}$

$k^2=\frac{2mE}{\hbar^2}$, so $E_n = \frac{k^2\hbar^2}{2m} = n^2\frac{\pi^2 \hbar^2}{2ma^2} (n=1, 2, 3 ...)$

$\psi(x)=B\sin{\frac{n\pi}{a}x} (0<x<a)$

According to the normalization condition, $\int_{-\infty}^{+\infty} {\left|\psi(x)\right|}^2 dx = \int_0^a B^2 \sin^2{\frac{n\pi}{a}x} dx = \frac{1}{2}aB^2 = 1$

thus, $\psi(x) = \sqrt{\frac{2}{a}} \sin{\frac{n\pi}{a}x}$

7 Use the theory of energy band to explain what is the insulator, semiconductor and conductor.

Metal conductivity is very high: the highest energy band filled by electrons, in metal, is not full，furthermore，the density of electron in the energy band is very high and it has the same magnitude as the density of the atom.

Insulator conductivity is very low: there is a large energy gap between the empty band and the full band, which suggests that very huge excitation energy is required to carry the electrons up over the gap into the next band. 

As for semiconductor, the energy gap is small. There can be a small, but not zero, density of electrons excited into the empty band. These electrons can easily carry a current, which would increase rapidly at high temperature.
